```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
customPalette = sns.color_palette("rocket_r", 8)

fig = plt.figure(figsize=(16,10))

ax = plt.axes(projection='3d')
ax.set_xlabel('Reach', fontsize=14)
ax.set_ylabel('Time Spent', fontsize=14)
ax.set_zlabel('Frequence', fontsize=14)
ax.set_xlim([0.,100.])
ax.set_ylim([0.,2.])
ax.set_zlim([0.,10.])

data = np.array([[0., 0., 0.], [1., 0.1, 1.], [5., 0.12, 1.2], [10., 0.15, 1.3], [15., 0.17, 1.4], [30, 0.2, 1.5], [70., 0.5, 3.], [100., 2., 10.]])

ax.plot3D(data[0:7, 0], data[0:7, 1], data[0:7, 2], 'gray',linestyle='--')
ax.plot3D(data[6:8, 0], data[6:8, 1], data[6:8, 2], 'gray',linestyle='dotted')

for idx, point in enumerate(data):
  ax.scatter(point[0],point[1],point[2], color=customPalette[idx])

plt.title('Goals\n', fontsize=18)
plt.show()
```
